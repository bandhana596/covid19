```python
import requests
from pandas.io.json import json_normalize
URL = "https://api.covid19india.org/data.json"
data = requests.get(url=URL).json()
df = json_normalize(data['statewise'])
print("Total Confirmed Cases: "+str(df[df.state == "Total"]['confirmed']))
print("Total Active Cases: "+str(df[df.state == "Total"]['active']))
print("Total Recovered Cases: "+str(df[df.state == "Total"]['recovered']))
print("Total Deceased Cases: "+str(df[df.state == "Total"]['deaths']))
print(df[['state','confirmed','active','recovered','deaths']].sort_values(by="confirmed", ascending=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np
state = df['state']
confirm = df['confirmed']
plt.xticks(rotation=90)
plt.title("State vs Confirmed-Cases",color='r',fontsize=20)
plt.xlabel("States",color='r',fontsize=20)
plt.ylabel("Number of Cases confirmed",color='r',fontsize=20)
plt.grid(True)
x=plt.bar(state,confirmed)
plt.show()
```

```python
active=df["active"]
state=df["state"]
plt.scatter(state,active)
plt.xticks(rotation=90)
plt.title("State vs Active-Cases",color='r',fontsize=20)
plt.xlabel("States",color='r',fontsize=20)
plt.ylabel("Number of Active Cases",color='r',fontsize=20)
plt.show()
```

```python
state=df["state"]
deaths=df["deaths"]
plt.plot(state,active)
plt.xticks(rotation=90)
plt.title("State vs deaths",color='r',fontsize=20)
plt.ylabel("Number of deaths",color='r',fontsize=20)
plt.show()
```
